```python
import pulp
model_1 = pulp.LpProblem("Problema 1", pulp.LpMinimize)
Xc = pulp.LpVariable("Xc", lowBound=1, upBound=250, cat="Integer")
Xp = pulp.LpVariable("Xp", lowBound=1, cat="Integer")
Xi = pulp.LpVariable("Xi", lowBound=1, upBound=300, cat="Integer")
Xm = pulp.LpVariable("Xm", lowBound=1, upBound=300, cat="Integer")
model_1 += 3 * Xc/100 + 8 * Xp/100 + 4 * Xi/100 + 2 * Xm / 100, "Min(3 * Xc/100 + 8 * Xp/100 + 4 * Xi/100 + 2 * Xm / 100)"
model_1 += 255 / 100 * Xc + 172 / 100 * Xp + 100 / 100 * Xi + 152 / 100 * Xm >= 2000, "restrictie 1 calorii"
model_1 += 5 / 100 * Xc + 25 / 100 * Xp + 9 / 100 * Xi + 0.8 / 100 * Xm >= 70, "restrictie 2 proteine"
model_1 += 48.3 / 100 * Xc + 0 / 100 * Xp + 4.7 / 100 * Xi + 33.8 / 100 * Xm >= 40, "restrictie 3 carbohidrati"
print(model_1)
status = model_1.solve()
print(f"Status solver: {pulp.LpStatus[status]}")
print(f"Solutiile: Xc={pulp.value(Xc)}, Xp={pulp.value(Xp)}, Xi={pulp.value(Xi)}, Xm={pulp.value(Xm)}, Min(3 * Xc/100 + 8 * Xp/100 + 4 * Xi/100 + 2 * Xm / 100)={pulp.value(model_1.objective)}")
```
