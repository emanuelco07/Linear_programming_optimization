```python
import random
import math

#definim functia obiectiv
def objective_function(Xc, Xp, Xi, Xm):
    return 3 * Xc/100 + 8 * Xp/100 + 4 * Xi/100 + 2 * Xm / 100

#definim restrictiile
def is_feasible(Xc, Xp, Xi, Xm):
    c1 = 2.55 * Xc + 1.72 * Xp + 1.00 * Xi + 1.52 * Xm >= 2000
    c2 = 0.05 * Xc + 0.25 * Xp + 0.09 * Xi + 0.008 * Xm >= 70
    c3 = 0.483 * Xc + 0 * Xp + 0.047 * Xi + 0.338 * Xm >= 40
    bounds = (1 <= Xc <= 250) and (1 <= Xp) and (1 <= Xi <= 300) and (1 <= Xm <= 300)
    return c1 and c2 and c3 and bounds

#parametrii simulated annealing
temp = 10.0          #temperatura initiala (mare => accepta multe erori)
cooling_rate = 0.999 #rata de racire (cat de repede scade T)
min_temp = 0.001     #pragul la care ne oprim
step_size = 10       #alterarea 
iterations = 0       #contor

#solutia initiala
current_sol = {"Xc": 150, "Xp": 150, "Xi": 150, "Xm": 150}
while not is_feasible(**current_sol):
    current_sol = {
        "Xc": random.randint(1, 250), "Xp": random.randint(1, 500),
        "Xi": random.randint(1, 300), "Xm": random.randint(1, 300)
    }

current_fitness = objective_function(**current_sol)
best_sol = current_sol.copy()
best_fitness = current_fitness

#bluca principala Simulated Annealing
while temp > min_temp:
    iterations += 1
    
    #cream o alterare
    new_sol = current_sol.copy()
    var_to_change = random.choice(list(new_sol.keys()))
    new_sol[var_to_change] += random.randint(-step_size, step_size)
    
    if is_feasible(**new_sol):
        new_fitness = objective_function(**new_sol)
        
        #calculam diferenta de energie (cost)
        # Delta E > 0 inseamna ca solutia obtinuta este mai slaba (cost mai mare)
        delta_e = new_fitness - current_fitness
        
        #regula de acceptare:
        #daca e mai buna (delta_e < 0), o acceptam mereu
        #daca e mai slaba, o acceptam cu o probabilitate e^(-delta_e / T)
        if delta_e < 0 or random.random() < math.exp(-delta_e / temp):
            current_sol = new_sol
            current_fitness = new_fitness
            
            #pastram separat cea mai buna solutie gasita vreodata
            if current_fitness < best_fitness:
                best_sol = current_sol.copy()
                best_fitness = current_fitness
    
    #racim sistemul
    temp *= cooling_rate

print(f"--- Rezultat Simulated Annealing (dupa {iterations} iteratii) ---")
print(f"Cea mai buna soluție: {best_sol}")
print(f"Cost minim absolut: {best_fitness}")
```
