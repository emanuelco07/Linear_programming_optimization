```python
import random

#definim functia obiectiv
def objective_function(Xc, Xp, Xi, Xm):
    return 3 * Xc/100 + 8 * Xp/100 + 4 * Xi/100 + 2 * Xm / 100

#definim restrictiile (returneaza true daca sunt respectate)
def is_feasible(Xc, Xp, Xi, Xm):
    c1 = 2.55 * Xc + 1.72 * Xp + 1.00 * Xi + 1.52 * Xm >= 2000
    c2 = 0.05 * Xc + 0.25 * Xp + 0.09 * Xi + 0.008 * Xm >= 70
    c3 = 0.483 * Xc + 0 * Xp + 0.047 * Xi + 0.338 * Xm >= 40
    
    #restrictii de domeniu
    bounds = (1 <= Xc <= 250) and (1 <= Xp) and (1 <= Xi <= 300) and (1 <= Xm <= 300)
    
    return c1 and c2 and c3 and bounds

#parametrii algoritmului
max_iterations = 10000
step_size = 5 #cu cat urcam sau coboram

#generam o solutie initiala, cautand-o pana obtinem una fezabila
current_sol = {"Xc": 150, "Xp": 150, "Xi": 150, "Xm": 150} #valori de start
while not is_feasible(**current_sol):
    current_sol = {
        "Xc": random.randint(1, 250),
        "Xp": random.randint(1, 500),
        "Xi": random.randint(1, 300),
        "Xm": random.randint(1, 300)
    }

current_fitness = objective_function(**current_sol)

#bucla principala hill climbing
for i in range(max_iterations):
    #cream o alterare (vecinatate)
    new_sol = current_sol.copy()
    var_to_change = random.choice(list(new_sol.keys()))
    new_sol[var_to_change] += random.randint(-step_size, step_size)
    
    #daca noua solutie e valida si mai buna (pretul este mai mic), o pastram
    if is_feasible(**new_sol):
        new_fitness = objective_function(**new_sol)
        if new_fitness < current_fitness:
            current_sol = new_sol
            current_fitness = new_fitness

print("--- Rezultat Hill Climbing ---")
print(f"Solutie: {current_sol}")
print(f"Cost minim gasit: {current_fitness}")
```
